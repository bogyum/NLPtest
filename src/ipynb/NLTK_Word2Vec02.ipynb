## 사전 훈련된 Word2Vec 임베딩

--------------------

- 구글에서 제공하는 사전 훈련된 Word2Vec 모델을 다운로드하여 사용합니다.
- 모델 다운로드 경로: https://drive.google.com/file/d/0B7XkCwpI5KDYNlNUTTlSS21pQmM/edit

```python
import gensim
import urllib.request
import matplotlib.pyplot as plt
```

```python
# 구글의 사전 훈련된 Word2Vec 모델을 로드.
# urllib.request.urlretrieve("https://s3.amazonaws.com/dl4j-distribution/GoogleNews-vectors-negative300.bin.gz", filename="../../resource/GoogleNews-vectors-negative300.bin.gz")
word2vec_model = gensim.models.KeyedVectors.load_word2vec_format('../../resource/GoogleNews-vectors-negative300.bin.gz', binary=True)
```

```python
print(word2vec_model.vectors.shape)
```

```python
print(word2vec_model.similarity('this', 'is'))
print(word2vec_model.similarity('post', 'book'))
```

```python
model_result = word2vec_model.most_similar('man', topn=10)
```

```python
print(model_result)
```

```python
def showGraph(bargraph):
    xtick = [item[0] for item in bargraph] # words
    ytick = [item[1] for item in bargraph] # 유사도
    plt.figure()
    mycolors = ['#06c2ac', '#c79fef', '#ff796c', '#aaff32', '#0485d1', '#d648d7', '#a5a502', '#d8dcd6', '#5ca904', '#fffe7a' ]
    plt.bar(xtick, ytick, color=mycolors)
```

```python
showGraph(model_result)
plt.show()
```

```python
model_result = word2vec_model.most_similar('frog', topn=10)
```

```python
print(model_result)
```
